```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import params as par
from matplotlib.ticker import AutoMinorLocator
from IPython.display import display, HTML
display(HTML("<style>.container { width:90% !important; }</style>"))
```

```python
data = pd.read_csv('data.csv')
```

```python
vtx = np.array(data['# vtx with two tracks of opposite curvature'])
x = data['XV (cm)'].to_numpy()
y = data['YV(cm)'].to_numpy()
z = data['ZV(cm)'].to_numpy()
px1 = data['Track1 PX(MeV) '].to_numpy()
py1 = data['Track1 PY(MeV) '].to_numpy()
pz1 = data['Track1 PZ(MeV) '].to_numpy()
px2 = data['Track2 PX(MeV) '].to_numpy()
py2 = data['Track2 PY(MeV) '].to_numpy()
pz2 = data['Track2 PZ(MeV) '].to_numpy()
```

```python
# DEFINING CUTS

cut1 = (z<120) & (z>-120) & (x<150) & (x>35) & (y<150) & (y>35)
```

```python
# COMPUTING ADDITIONAL QUANTITIES

r = np.sqrt(x*x+y*y+z*z)

E1 = par.getE(par.m_pi, px1, py1, pz1)
E2 = par.getE(par.m_pi, px2, py2, pz2)
inv_mass = par.invmass(E1+E2, px1+px2, py1+py2, pz1+pz2)

E_K = np.repeat(np.sqrt(par.m_K0**2 + par.p_K0**2), len(x))
p_K = par.getComponents(np.repeat(par.p_K0, len(x)), x, y, z)
missing_mass = par.invmass(E_K - E1 - E2, p_K[0] - px1 - px2, p_K[1] - py1 - py2, p_K[2] - pz1 - pz2)
missing_mom = par.norm(p_K[0] - px1 - px2, p_K[1] - py1 - py2, p_K[2] - pz1 - pz2)
```

```python
plt.tick_params(which='major', length=7, direction = 'in')
plt.tick_params(axis = 'both', which='minor', length=3, color='k', direction = 'in')
plt.minorticks_on()
plt.ticklabel_format(axis='y', style='sci', scilimits=(0,3), useMathText = True)
plt.hist(r[cut1], histtype = 'step', range = (0,200), bins = 10)
plt.hist(r, histtype = 'step',range = (0,200), bins = 10)
plt.xlim(0,r.max());
```

```python
pos = np.arange(0.,30, 0.1)
ratio_L = (par.l_L/par.l_S) * (np.exp(-pos/par.l_S)/np.exp(-pos/par.l_L))
like_L = 1/(1+ratio_L)

ratio_S = (par.l_S/par.l_L) * (np.exp(-pos/par.l_L)/np.exp(-pos/par.l_S))
like_S = 1/(1+ratio_S)

plt.plot(pos, like_L, label = 'likelihood KL')
plt.plot(pos,like_S, label = 'likelihood KS')
plt.legend();
```

```python
sliced = data.iloc[6:10]
sliced
```

```python
theta = []
phi = []
r = []

N = sliced['#'].values
tolerance = np.pi/15

for i in N:
    theta.append(par.getTheta(sliced['XV (cm)'][i], sliced['YV(cm)'][i], sliced['ZV(cm)'][i]))
    phi.append(par.getPhi(sliced['XV (cm)'][i], sliced['YV(cm)'][i], sliced['ZV(cm)'][i])) 
    r.append(par.norm(sliced['XV (cm)'][i], sliced['YV(cm)'][i], sliced['ZV(cm)'][i]))
 

for i in range(len(N)):
    for j in range(len(N)):
        if i != j and j>i:
            if (-tolerance < abs(theta[i] - theta[j]) < +tolerance):
                print(f'Theta checked {i+1} {j+1}')
                if (np.pi-tolerance < abs(phi[i] - phi[j]) < np.pi+tolerance):
                    print(f'Phi checked {i+1} {j+1}')
                    if (r[i] <= r[j]):
                        print(f'{i + N[0]} is a KS')
                        print(f'{j + N[0]} is a KL')
                    elif (r[i] > r[j]):
                        print(f'{i + N[0]} is a KL')
                        print(f'{j + N[0]} is a KS')
            else:
                print('No kaon to tag')
```

```python
from matplotlib.patches import Rectangle

event = N
fig, ax = plt.subplots(1, 2,figsize=(10, 5))


for i in event:

    col = (np.random.random(), np.random.random(), np.random.random())
    
    V = [x[i],y[i]]
    origin = [0,0] # origin point

    ax[0].quiver(*origin, V[0], V[1], color = col , angles='xy', scale_units='xy',
               scale=1, headlength = 2, headaxislength = 2, width = 0.005, ls = '--')
    
    track1 = [px1[i],py1[i]]
    ax[0].quiver(*V, track1[0], track1[1], color=col, angles='xy', scale_units='xy', 
               scale=1, headlength = 2, headaxislength = 2, width = 0.003)
    
    track2 = [px2[i],py2[i]]
    ax[0].quiver(*V, track2[0], track2[1], color=col, angles='xy', scale_units='xy', 
               scale=1, headlength = 2, headaxislength = 2, width = 0.003)

    ax[0].add_patch(Rectangle((-150, -150), 300, 300,
             edgecolor = 'cyan',
             facecolor = 'blue',
             fill=False,
             lw=1))
    
    V = [z[i],y[i]]
    origin = [0,0] # origin point

    ax[1].quiver(*origin, V[0], V[1], color = col , angles='xy', scale_units='xy',
               scale=1, headlength = 2, headaxislength = 2, width = 0.005, ls = '--')
    
    track1 = [pz1[i],py1[i]]
    ax[1].quiver(*V, track1[0], track1[1], color=col, angles='xy', scale_units='xy', 
               scale=1, headlength = 2, headaxislength = 2, width = 0.003)
    track2 = [pz2[i],py2[i]]
    ax[1].quiver(*V, track2[0], track2[1], color=col, angles='xy', scale_units='xy', 
               scale=1, headlength = 2, headaxislength = 2, width = 0.003)
    
    ax[1].add_patch(Rectangle((-120, -150), 240, 300,
         edgecolor = 'cyan',
         facecolor = 'blue',
         fill=False,
         lw=1))
    

ax[0].set_xlim([-200, 200])
ax[0].set_ylim([-200, 200])
ax[1].set_xlim([-200, 200])
ax[1].set_ylim([-200, 200])
ax[0].set_facecolor('black')
ax[1].set_facecolor('black')
```
