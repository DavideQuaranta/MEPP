```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'png'
```

Evaluate the integral I = $\int_{-\infty}^{+\infty} e^{-\frac{x^2}{2}}dx$ using two Monte Carlo methods:
   1. Miss or hit
   2. The mean method
   3. From the obtained value, considering that I = $\sqrt{2\pi}$ , get an estimate of the
       irrational number $\pi$.

```python
def func(x):
    return np.exp(-x**2/2)
```

```python
x = np.arange(-20,20, 0.01)
plt.title(r'Plot of the function $e^{-\frac{x^2}{2}}$')
plt.plot(x, func(x))
plt.show()
```

### 1) Miss or Hit

```python
# define the boundaries of the rectangle
# we want to extract numbers from

a = 5
h = 1.0 + 0.01

# Number of numbers to extract
N = 1000000

rand_x = np.random.uniform(-a, a, size = N)
rand_y = np.random.uniform(0, h, size = N)

y_hit = [rand_y[i] for i in range(N) if (rand_y[i] <= func(rand_x[i]))]
x_hit = [rand_x[i] for i in range(N) if (rand_y[i] <= func(rand_x[i]))]

I = (len(y_hit)/N) * (h*2*a)
print(f'I (hit or miss) = {I:.4f}')
print(f'I (true) = {np.sqrt(2*np.pi):.4f}')
print(f'Error(%) = {100*(abs(np.sqrt(2*np.pi)-I)/I):.4f}')
```

```python
plt.scatter(rand_x, rand_y, zorder = -1, color = 'orange', alpha = 0.01, edgecolor = 'orange')
plt.scatter(x_hit, y_hit, zorder = 0, color = 'cornflowerblue', alpha = 0.01, edgecolor = 'navy')
plt.plot(x, func(x), zorder = 1 , color = 'red')
plt.xlim(-a-1,a+1)
plt.show()
```

### 2) Mean Method

```python
I_mean = (2*a/N)*np.sum(func(rand_x))

print(f'I (mean method) = {I_mean:.4f}')
print(f'I (true) = {I:.4f}')
print(f'Error(%) = {100*(abs(np.sqrt(2*np.pi)-I_mean)/I_mean):.4f}')
```

### 3) Pi Estimate

```python
Pi_HM = I**2/2
Pi_Mean = I_mean**2/2

print(f'Pi (Hit or Miss) = {Pi_HM:.4f}')
print(f'Error(%) = {100*(abs(np.pi-Pi_HM)/Pi_HM):.4f}')
print(f'Pi (Mean Method) = {Pi_Mean:.4f}')
print(f'Error(%) = {100*(abs(np.pi-Pi_Mean)/Pi_Mean):.4f}')
```
