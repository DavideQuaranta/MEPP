```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def f1(x):
    return 1/x

low_lim = 0.1
high_lim = 0.92
```

```python
length = 40000

N = 10000

rand_E = np.random.uniform(low_lim, high_lim, size = length)
rand_Y = np.random.uniform(0, f1(low_lim), size = length)

hit = rand_Y <= f1(rand_E)



plt.scatter(rand_E[hit][0:N], rand_Y[hit][0:N], color = 'navy', alpha = 0.05, s=20)
plt.scatter(rand_E[hit][0:N], rand_Y[hit][0:N], facecolors='none', edgecolor = 'blue', alpha = 0.1, s=20)

plt.scatter(rand_E[hit == False][0:N], rand_Y[hit == False][0:N], color = 'orange',alpha = 0.1, s=20)
plt.scatter(rand_E[hit == False][0:N], rand_Y[hit == False][0:N], facecolors='none', edgecolor = 'darkorange', alpha = 0.1, s=20)
x = np.arange(low_lim, high_lim, 0.01)
plt.plot(x, f1(x), color = 'red')
plt.xlabel('Energy [MeV]')
plt.ylabel('f1(E)')
plt.show()
```

```python
E = rand_E[hit][0:N]

smear_E = [np.random.normal(energy, 0.1*np.sqrt(energy)) for energy in E]

for i in range(N):
    if smear_E[i] <= 0:
        smear_E[i] = 0

n, bins, patches = plt.hist(E, bins = 100,range = (0,3), histtype = 'step', label = 'True', density = True)
plt.hist(smear_E, bins = 100,range = (0,3), histtype = 'step', label = r'Smeared 10%', density = True)

conv_int_x, conv_int_y = np.load('res10_data.npy')

plt.plot(conv_int_x, conv_int_y/(2.247126695273289), color = 'red', label = 'Conv. Integral')

plt.legend()
plt.show()
```
